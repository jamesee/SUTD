## **What is an algorithm? What is its relation to data structures?**

An **algorithm** is any well-defined computational procedure that takes some value, or set of values, as **input** and produces some value, or set of values as **output**. An algorithm is thus a *sequence of computational steps* that transform the input into the output.

We can also view an algorithm as a tool for solving a well-specified computational problem. The statement of the problem specifies in general terms the desired input/output relationship.

**Data structures** are a key part of computer algorithms, as they allow us to perform data operations efficiently. A wise choice of data structures can improve the performance of a program/algorithm.

**EXAMPLE** *Sorting Problem*: 

We might need to sort a sequence of numbers into ***increasing order***. This is something that arises frequently in practice. For instance, we could be trying to locate the median exam score of students. which requires finding the middle ordered score in a sorted sequence of scores.

We can formally define the **sorting problem**:

> **Input**: A sequence of $n$ numbers $\langle a_1, a_2, ... , a_n \rangle$
> 
> **Output**: A reordering of $\langle a'_1, a'_2, ... , a'_n \rangle$ of the input sequence such that $a'_1 \leq a'_2 \leq ... \leq a'_n$.

## **Two Aspects of Algorithm Design: Correctness and Efficiency**

An algorithm is said to be **correct**, if for every instance, it always returns the desired output.

An algorithm is **efficient** if it executes as quickly as possible.

### **Optimising the Travel Time of a Robot Arm**

Presume you have a robot arm equipped with a soldering iron. The robot arm's job is to solder the required contact points so that the electronic circuit board will be completed.

We therefore seek to find an **optimal tour** of the different contact points such as to minimise the travel distance of the robotic arm (since this means we can quickly move on to the next circuit board!).

### **A Correct Solution Brute-Force/Exhaustive Search**

A **correct** algorithm to the *Travelling Salesman Problem* (TSP) as presented in the previous section is to perform a **brute-force/exhaustive search of all possible orderings of the points**, and then selecting the one which minimises the total length.

Since all possible orderings are considered, we are guaranteed to end up with the shortest possible tour.

Consider the following: When the number of points is small, the total amount of time required for exhaustive search is *negligible*. However, a simple increment to >20 points, and this time increases. **In fact, with $n$ points, we have to check $n!$ permutations!**

Thus in the design of algorithms, it is important to consider both **correctness** and **efficiency**.

## **Analysing Algorithms**

The analysis of algorithms means predicting the resources that the algorithm requires. While resources such as memory (*space complexity*), communication bandwidth, or computer hardware are of primary concern, most often it is **computational time** (*time complexity*)that we want to measure.

The complexity of an algorithm is often expressed as a function of the **input size**. For example:
- Sorting: number of input items
- Multiplication: total nuber of bits or digits
- Graph algorithms: number of nodes & edges

### **Types of Analysis**

#### Worst Case Running Time

- The worst case running time of an algorithm is an **upper bound** on the running time for any input. 
- Knowing it gives us a guarantee that the algorithm will never take any longer.

#### Average Case

- The expected behaviour when the input is randomly drawn from a given distribution. The average case running time is an estimate of the running time for an **expected** input. 
- Its computation entails knowing all possible input sequences, the probability distribution of occurence of these sequences, and the running times for the invididual sequences. 
- We often assume that all inputs of a given size are equally likely.

#### Amortised Analysis

- Time required to perform a sequence of related operations is averaged over all the operations performed.

### **2. Dynamic Programming**

Dynamic Programming is mainly an optimisation over plain recursion. Wherever we see a solution that was written using recursion that has repeated calls for the same inputs, we can optimise it using Dynamic Programming. 

The key idea is ***memoisation***, which is to store the results of subproblems, so that we do not have to re-compute them when needed later. This simple optimisation can significantly reduce the time complexity from exponential to polynomial time. The Fibonacci Sequence is a classic problem that can benefit from memoisation.

### **3. Greedy Algorithms**

Greedy is an algorithmic paradigm that follows the problem-solving heuristic of making the **locally optimal** choice at each stage. While for many problems, a greedy strategy does not produce an optimal solution, the greedy heuristic can yield locally optimal solutions that *approximate* a globally optimal solution in a reasonable amount of time.
