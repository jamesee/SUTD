# **Errors are inevitable**

Errors are bound to occur in your code. This is especially so, when someone else uses your code in an unexpected way.

We can thus use **error handling** to attempt to plan for possible errors.

There are (at least) two distinguishable kinds of errors: [syntax errors](https://docs.python.org/3/tutorial/errors.html#syntax-errors) and [exceptions](https://docs.python.org/3/tutorial/errors.html#exceptions).

**Syntax errors** are also known as parsing errors, are one of the most common errors you will encounter when you are learning Python. The use of integrated development environments (IDEs) such as Colab, which have code completion or syntax checkers helps to catch some of these errors.

**Exceptions**. Even syntactically correct code can result in execution errors. Errors detected during execution are called exceptions.

# **`try`, `except`, `else` and `finally` to handle errors**

- `try` lets you ***test*** a block of code for errors. If an error is detected, control flow jumps to the `except` block immediately.
- `except` lets you ***handle*** the error
- `else` lets you **execute code, only if no exceptions were raised in the try block**
- `finally` lets you ***execute code, regardless*** of the try and except blocks

The following example shows how `try` and `except` are used: 
- The print statement (line 2) is the code that is to be tested
- Since `x` was not declared beforehand, this will result in an error, which gets caught in the `try` block. 
- Once the error has been caught, we go to the `except` block to handle the error. Here in line 4, we simply print that an exception has occured. 

```python
try:
  print(x)
except:
  print("An exception occured")
```

In some instances, we want our code to do something, but there may exist an error. If there is no error that is thrown, we want to take the outputs of that processing to do something else. In this case, we use the `else` keyword. The following example shows how this is done.

```python
try:
  # Let's presume there is a possibility for error in this block
  result = "Singa" + "pore"
except:
  print("Check that the values are both strings!")
else:
  print(f"Welcome to {result}!")
```

The following two code examples show how `finally` works. In the first code block, x has not been defined, and in the second code block y has been defined, but both examples will execute the `finally` block.

```python
# x is not defined
try:
  print(x)
except:
  print("An exception occured")
finally:
  print('Executed regardless of try and except')
```

```python
# y is defined
y = 10
try:
  print(y)
except:
  print("An exception occured")
finally:
  print('Executed regardless of try and except')
```

# **Built-In Exceptions**

Python has a list of [built-in exceptions](https://docs.python.org/3/library/exceptions.html#bltin-exceptions). In fact, we saw an example of this earlier, a `ZeroDivisionError`:

Using built-in exceptions allows you to quickly identify some common sources of error such as the `ZeroDivisionError`.

From the previous example, instead of a generalised except, we can now catch **specific errors**, and provide handle the error accordingly--in this case, to give diagnostic feedback to indicate what went wrong.

```python
try:
  print(x)
except NameError:
  print("Variable x is not defined")
```

# **Catching Different Exceptions**

Since a named exception has known causes, catching that specific error can allow you to provide specific remedies.

```python
try:
  f = open('testfile', 'w') # changing from 'w' to 'r' will generate an OSError
  f.write("write out a line of text")
except TypeError:
  print("there was a TypeError")
except OSError:
  print("there was an OSError")
except:
  print("caught all other exceptions")
finally:
  print("this code always runs")
  f.close()
```

# **Example**

Let's presume we want to create a game in which a user has to input his guess for an integer.

```python
import random

def guess_the_int():
  num = random.randint(0, 9)

  while True:
    try:
      result = int(input("Please provide a number between 0-9: "))
    except:
      print("Oops, that's not a number")
      continue
    else:
      if num == result:
        break
      print("WRONG! Try again!")
    finally:
      print("--------------------------")

  print(f"Nice guess, the number is {num}")


guess_the_int()
```

# **Raising an Exception**

Here is a trivial example of raising an exception:

```python
x = -2

if x < 0:
  raise Exception("Sorry, no negative numbers allowed")
```

Why do we want to raise an exception? In the previous example, the `int()` function expects that a number is given to it. Since the function only converts any number to an integer, it does not know how to handle things like strings. By raising an exception, the exception can be handled and as per that example, we informed the user that the input was not a number, and requested for a number input. This allowed the program to continue working and not stop. Great!
