# **DEBUGGING BASICS**

## **Why Debug?**
Even the best programmers are unable to write code once and have it run perfectly all the time. Bugs will exist, and therefore debugging is an essential skill to develop for all programmers.

Let's examine some basic debugging techniques that can be used in programming.

---
# **Simple Strategies**

## **Coding in Small Segments**

Writing your code in small segments at a time allows you to test whether your code is working the way you expect it to. This prevents errors from propagating to other parts of your code, and makes it easier to spot them.

If the program has already been written, apply a divide and conquer strategy to chop up the code into smaller sections to debug.

## **Trial and Error**

Identifying a bug often requires you to investigate the underlying problem, formulate some hypotheses on what went wrong, and testing it out. If you systematically eliminate the possibilities, then you're likely to weed out those nasty bugs.

## **Compare with Parallel Examples**

If it's hard to identify the error, you can find a parallel small snippet of code online that tries to perform the same function to understand how it should be done. This may provide a clue as to what went wrong with your code.

In fact, there are many "cookbooks" that have been written for the various programming languages that give you these snippets that tackle many common problems in writing a program.

## **Ask for Help**

There's no shame in admitting that you sometimes cannot see where the bug is and ask for help. Having fresh eyes to look at your code very often can help in identifying the bugs in it. In a similar vein, you may also take a break from working on that code, and come back to it later.

In fact, an entire community has been built around asking for help on programming: [Stack Overflow](https://stackoverflow.com/).

# **Reading Error Messages**

Error messages are a key aspect of debugging. While integrated development environments (IDEs) such as this Google Colab are able to highlight *some* potential syntax errors, and use code suggestions to help with writing your code, syntax-error free code may still contain **runtime errors**. When Python is able to detect such runtime errors, it complains and prints out an **error message**.

The following code is one such example.

## **Try interpreting this error message!**

````python
def divide(num):
  print(num)
  n = int(num)
  return 100 / n

print(divide(0))
````

## **Error traces don't always show the right line of the error**

````python
def divide(num):
  print(num)
  n = int(num
  return 100 / n

print(divide(0))
````

The error message above seems to suggest that the error occurred at line 4 of the code--**but it's not**!

By reading the error message, a `SyntaxError`, and on examining line 4, line 4 appears to be perfectly legitimate. In such situations, you may wish to examine the neighbouring lines of code for potential syntax errors. In this example, there is a **missing closing parentheses** in line 3 which the Python interpreter expects. This leads the error that overflows to line 4.

# **Print and Check**

A simple but powerful method for debugging is to simply print variables out to check if their values are as expected or not.

While the error trace helps in narrowing down where the error is, it is not always diagnostic. By performing `print` we can better check to see what went wrong.

Screenshot 2021-06-14 at 4.28.47 PM.png

````python
def divide(num):
  print(num)
  n = int(num)
  # since the error happened on `return 100 / n`, 
  # it's a good idea to check what happened before that
  print(n) # add this print statement 
  return 100 / n

print(divide(0))
````

The first 0 comes from `print(num)`, and the second 0 comes from `print(n)` which we added to figure out what went wrong.

We can now clearly see that we have passed a value of `n = 0` to `return 100 / n`, a clear case of an programming and arithmetic error. 

In such cases, we may want to **handle the error** gracefully instead of allowing the programme to terminate.

# **Assert and Check**

While simple print statements are powerful, they add additional code into your programme which is not part of the functionality. The alternative is of course to perform housekeeping, and ensure that you remove all of this code.

A simpler way is to use the `assert` **keyword**. An assert statement takes the format: `assert <Boolean statement>`, for example: `assert 1 == 0` would throw an `AssertionError`. When an assertion evaluate to `True`, the program continues normal execution.

> tip.png
Pro Tip: `assert` is a keyword, and not a function! Don't use it this way `assert()`.

While this may appear to also include additional code that is not part of the required functionality, the execution of a Python script under *optimised mode* would instruct the Python interpreter to **completely ignore asserts**.

**Running Python in optimised mode**:
```
python -O my_script.py
```

Here's an example of how to use `assert`:

````python
def divide(num):
  # We added an assert here to ensure that that num is always an integer
  assert type(num) == int
  n = int(num)
  return 100 / n

print(divide(10))
print(divide("Ten"))
````

# **Other methods**

The following methods are beyond the scope of the course, but also worth mentioning. 

## **Logging**
In professional programming, most languages or frameworks support the debugging process by providing a means to log messages such as information, warnings, or errors into a log file. A log file is much easier to filter and review as compared to the console output. Python provides the [`logging`](https://docs.python.org/3/library/logging.html) module to do this.

## **Debugger**
Many IDEs provide a means to attach a debugger to insert breakpoints, inspect the state of the program and to execute the code in a controlled fashion. For example, using a debugger, you may step through your code line-by-line and examine how the variables change as you do so. Python provides the debugger module [`pdb`](https://docs.python.org/3/library/pdb.html), which provides this sort of functionality.
